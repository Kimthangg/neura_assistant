```python
from sentence_transformers import SentenceTransformer
import torch
import os
base_path = "./services/embedding_model/models--hiieu--halong_embedding/snapshots"
snapshots = os.listdir(base_path)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
if not snapshots:
    model = SentenceTransformer("hiieu/halong_embedding", cache_folder = "./", device=device) 
else:
    model_path = os.path.join(base_path, snapshots[0])
    model = SentenceTransformer(model_path, device=device)
# Define query and documents
query = "liên hệ với ftech"

# Encode query and documents
query_embedding = model.encode(query)
print(query_embedding.tolist())

```
